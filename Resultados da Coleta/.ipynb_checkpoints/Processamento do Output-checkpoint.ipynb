```python
#Antes de processar eu eliminei as colunas que estavam sobrando.
import pandas as pd
import numpy as np
import datetime
```

```python
df = pd.read_excel('OUTPUTS JUNTOS.xlsx')
df = df.reset_index(drop = True)
print('Número inicial de olhos: ', df.shape[0])
df1 = df
```

```python
#Reordenando as Colunas para modo de trabalho normal

cols = ['Nome', 'Registro', 'Data da cirurgia', 'Idade', 'Olho', 'Dioptria', 'Marca da Lente', 'Modelo da Lente', 'Data Pré-op',
       'Esférico Pré-op', 'Cilindro Pré-op', 'Eixo Pré-op', 'Acuidade Pré-op', 'AR Pré-op', 'Data Pós-op', 'Esférico Pós-op', 
        'Cilindro Pós-op', 'Eixo Pós-op', 'Acuidade Pós-op', 'AR Pós-op']
df = df[cols]
```

```python
#Corrigindo o Erro da Transposição de alguns campos em olhos esquerdos, já filtrando alguns erros.
removed1 = pd.DataFrame(columns = cols)

for index in range(df.shape[0]):
    if index%50 == 0:
        print(index, ' de ', df.shape[0], ' fichas.')
    val = df['Esférico Pós-op'][index]
    if isinstance(val, int) or isinstance(val, float):
        #######################
        if val > 1000:
            esfPre = df['Cilindro Pré-op'][index]
            cilPre = df['Eixo Pré-op'][index]
            eixoPre = df['Acuidade Pré-op'][index]
            acuPre = df['Esférico Pré-op'][index]
            esfPos = df['Cilindro Pós-op'][index]
            cilPos = df['Eixo Pós-op'][index]
            eixoPos = df['Acuidade Pós-op'][index]
            acuPos = df['Esférico Pós-op'][index]
            
            df['Esférico Pré-op'][index] = esfPre
            df['Cilindro Pré-op'][index] = cilPre
            df['Eixo Pré-op'][index] = eixoPre
            df['Acuidade Pré-op'][index] = acuPre
            df['Esférico Pós-op'][index] = esfPos
            df['Cilindro Pós-op'][index] = cilPos
            df['Eixo Pós-op'][index] = eixoPos
            df['Acuidade Pós-op'][index] = acuPos
            continue
            
    if isinstance(val, str):
        val = val.strip()
        if ',' in val:
            val = val.replace(',', '.')
            val = val.replace(';', '.')
        #######################
        if '/' in val:
            esfPre = df['Cilindro Pré-op'][index]
            cilPre = df['Eixo Pré-op'][index]
            eixoPre = df['Acuidade Pré-op'][index]
            acuPre = df['Esférico Pré-op'][index]
            esfPos = df['Cilindro Pós-op'][index]
            cilPos = df['Eixo Pós-op'][index]
            eixoPos = df['Acuidade Pós-op'][index]
            acuPos = df['Esférico Pós-op'][index]
            
            df['Esférico Pré-op'][index] = esfPre
            df['Cilindro Pré-op'][index] = cilPre
            df['Eixo Pré-op'][index] = eixoPre
            df['Acuidade Pré-op'][index] = acuPre
            df['Esférico Pós-op'][index] = esfPos
            df['Cilindro Pós-op'][index] = cilPos
            df['Eixo Pós-op'][index] = eixoPos
            df['Acuidade Pós-op'][index] = acuPos
            continue
            
        #######################
        if val.lower() == 'plano' or val.lower() == 'pl':
            df['Esférico Pós-op'][index] = 0
            esfPre = df['Cilindro Pré-op'][index]
            cilPre = df['Eixo Pré-op'][index]
            eixoPre = df['Acuidade Pré-op'][index]
            acuPre = df['Esférico Pré-op'][index]
            esfPos = df['Cilindro Pós-op'][index]
            cilPos = df['Eixo Pós-op'][index]
            eixoPos = df['Acuidade Pós-op'][index]
            acuPos = df['Esférico Pós-op'][index]
            
            df['Esférico Pré-op'][index] = esfPre
            df['Cilindro Pré-op'][index] = cilPre
            df['Eixo Pré-op'][index] = eixoPre
            df['Acuidade Pré-op'][index] = acuPre
            df['Esférico Pós-op'][index] = esfPos
            df['Cilindro Pós-op'][index] = cilPos
            df['Eixo Pós-op'][index] = eixoPos
            df['Acuidade Pós-op'][index] = acuPos
            continue
        if "+" in val:
            if val.index('+') > 0:
                try:
                    val = float(val[0:val.index('+')])
                    df['Esférico Pós-op'][index] = val
                    continue
                except Exception as e:
                    print(e)
                    print('Removendo ', df['Nome'][index], ' Esférico: ', df['Esférico Pós-op'][index])
                    removed1 = removed1.append(df.loc[index, :])
                    df = df.drop([index])
                    continue
        if " " in val:
            try:
                val = float(val[0:val.index(" ")])
                df['Esférico Pós-op'][index] = val
                continue
            except Exception as e:
                print(e)
                print('Removendo ', df['Nome'][index], ' Esférico: ', df['Esférico Pós-op'][index])
                removed1 = removed1.append(df.loc[index, :])
                df = df.drop([index])
                continue
        try:
            val = float(val)
            df['Esférico Pós-op'].loc[index] = val
            continue
        except Exception as e:
            print(e)
            print('Removendo ', df['Nome'][index], ' Esférico: ', df['Esférico Pós-op'][index])
            removed1 = removed1.append(df.loc[index, :])
            df = df.drop([index])
            continue
        if not val.isdigit():
            print('Removendo ', df['Nome'][index], ' Esférico: ', df['Esférico Pós-op'][index])
            removed1 = removed1.append(df.loc[index, :])
            df = df.drop([index])
            continue

#df = df.reset_index(drop = True)
df2 = df.copy()
removed1.to_excel('Removidos1.xlsx')
```

```python
print('Número de olhos após realizar correção do erro e fazer limpeza superficial do esférico pós-op: ', df.shape[0])
```

```python
#Limpeza dos dados essenciais em falta: datas, esféricos pós-op, dioptrias, modelos de lente, acuidade pós-op
df = df.dropna(subset=['Data da cirurgia', 'Dioptria', 'Modelo da Lente', 'Data Pré-op', 'Data Pós-op', 'Esférico Pós-op',
                      'Acuidade Pós-op'])
df = df.reset_index(drop=True)
df3 = df

print('Número de olhos após remover os dados faltantes: ', df.shape[0])
```

```python
#Removendo quem não tem acuidade pós-operatória > 20/40
for index in range(df.shape[0]):
    val = df['Acuidade Pós-op'][index]
    
    #Tratando as 5 exceções que seriam perdidas sem motivo:
    if val == '2/20':
        val = 2020;
        df['Acuidade Pós-op'][index] = val
        continue;
    if val == '20/25+2,5':
        val = 2025;
        df['Acuidade Pós-op'][index] = val
        continue;
    if val == '20/15P':
        val = 2015;
        df['Acuidade Pós-op'][index] = val
        continue;
    ### FIM DO TRATAMENTO DAS EXCEÇÕES
    
    if isinstance(val, str):
        if '/' in val:
            try:
                val = int(val[0:2] + val[3:])
                df['Acuidade Pós-op'][index] = val
                continue
            except:
                #print('Apagando: ', val)
                df = df.drop([index])
                continue
        elif val.isdigit():
            val = int(val)
            df['Acuidade Pós-op'][index] = val
        else:
            #print('Apagando: ', val)
            df = df.drop([index])
            continue
    

df['Acuidade Pós-op'] = pd.to_numeric(df['Acuidade Pós-op'], errors = 'coerce')
df = df.dropna(subset=['Acuidade Pós-op'])
df = df.reset_index(drop=True)

for index in range(df.shape[0]):
    if df['Acuidade Pós-op'][index] > 2040:
        df = df.drop([index])
    elif df['Acuidade Pós-op'][index] < 2000:
        df = df.drop([index])

df = df.reset_index(drop=True)

print('Número de olhos com acuidade pós-operatória < 20/50: ', df.shape[0])
df4 = df
```

```python
#Tratando as datas para DateTime, removendo os erros.
df['Data da cirurgia'] = pd.to_datetime(df['Data da cirurgia'], dayfirst=True, errors='coerce')    
df['Data Pré-op'] = pd.to_datetime(df['Data Pré-op'], dayfirst=True, errors='coerce')
df['Data Pós-op'] = pd.to_datetime(df['Data Pós-op'], dayfirst=True, errors='coerce')

df = df.dropna(subset=['Data da cirurgia', 'Data Pré-op', 'Data Pós-op'])

df = df.reset_index(drop=True)
print('Número de olhos com datas corretas: ', df.shape[0])
df5 = df
```

```python
#Removendo os pacientes fora do intervalo pós-operatório de 1 a 6 meses
for index in range(df.shape[0]):
    delta = (df['Data Pós-op'][index]-df['Data da cirurgia'][index]).days
    if delta < 30 or delta > 180:
        df = df.drop([index])

df = df.reset_index(drop=True)
        
print('Número de olhos com intervalo apropriado: ', df.shape[0])
df6 = df
    
#Removendo apenas os pacientes de pós-operatório de menos de 1 mês
df = df5

for index in range(df.shape[0]):
    delta = (df['Data Pós-op'][index]-df['Data da cirurgia'][index]).days 
    if delta < 30:
        df = df.drop([index])

df = df.reset_index(drop=True)
        
print('Número de olhos com intervalo maior que 1 mês: ', df.shape[0])
df7= df
#Removendo apenas os pacientes de pós-operatório de mais de 6 meses
df = df5

for index in range(df.shape[0]):
    delta = (df['Data Pós-op'][index]-df['Data da cirurgia'][index]).days 
    if delta > 180:
        df = df.drop([index])

df = df.reset_index(drop=True)
        
print('Número de olhos com intervalo menor que 6 meses: ', df.shape[0])
df8 = df
#Quantidade de olhos perdidos porque o segundo olho ficou com intervalo muito curto
#Quantidade de olhos perdidos porque o primeiro olho ficou com intervalo muito longo

df = df6
```

```python
#Tratamento das Dioptrias
for index in range(df.shape[0]):
    val = df['Dioptria'][index]
    if isinstance(val, str):
        val = val.strip()
        val = val.replace(',', '.')
        val = val.replace(';', '.')
        if '+' in val:
            df['Dioptria'][index] = val[0:val.index('+')]
        if '/' in val:
            df['Dioptria'][index] = val[0:val.index('/')]
        if ' ' in val:
            df['Dioptria'][index] = val[0:val.index(' ')]

df['Dioptria'] = pd.to_numeric(df['Dioptria'], errors='coerce')

df = df.dropna(subset = ['Dioptria'])
df = df.reset_index(drop=True)
        
print('Número de olhos com dioptrias apropriadas: ', df.shape[0])
```

```python
#Tratamento dos Esféricos
df['Esférico Pré-op'].value_counts()
for index in range(df.shape[0]):
    val = df['Esférico Pré-op'][index]
    if isinstance(val, str):
        val = val.strip()
        val = val.replace(',', '.')
        val = val.replace(';', '.')
        if '+' in val:
            df['Esférico Pré-op'][index] = val[0:val.index('+')]
        if '/' in val:
            df['Esférico Pré-op'][index] = val[0:val.index('/')]
        if ' ' in val:
            df['Esférico Pré-op'][index] = val[0:val.index(' ')]

df['Esférico Pré-op'] = pd.to_numeric(df['Esférico Pré-op'], errors='coerce')

#df = df.dropna(subset = ['Esférico Pré-op'])
df = df.reset_index(drop=True)
        
print('Número de olhos com esféricos pré-ops apropriados: ', df.shape[0])

df['Esférico Pós-op'].value_counts()
for index in range(df.shape[0]):
    val = df['Esférico Pós-op'][index]
    if isinstance(val, str):
        val = val.strip()
        val = val.replace(',', '.')
        val = val.replace(';', '.')
        if '+' in val:
            df['Esférico Pós-op'][index] = val[0:val.index('+')]
        if '/' in val:
            df['Esférico Pós-op'][index] = val[0:val.index('/')]
        if ' ' in val:
            df['Esférico Pós-op'][index] = val[0:val.index(' ')]

df['Esférico Pós-op'] = pd.to_numeric(df['Esférico Pós-op'], errors='coerce')

df = df.dropna(subset = ['Esférico Pós-op'])
df = df.reset_index(drop=True)
        
print('Número de olhos com esféricos pós-ops apropriados: ', df.shape[0])
```

```python
#Tratamento dos Cilindros
df['Cilindro Pré-op'].value_counts()
for index in range(df.shape[0]):
    val = df['Cilindro Pré-op'][index]
    if isinstance(val, str):
        val = val.strip()
        val = val.replace(',', '.')
        val = val.replace(';', '.')
        if '+' in val:
            df['Cilindro Pré-op'][index] = val[0:val.index('+')]
        if '/' in val:
            df['Cilindro Pré-op'][index] = val[0:val.index('/')]
        if ' ' in val:
            df['Cilindro Pré-op'][index] = val[0:val.index(' ')]

df['Cilindro Pré-op'] = pd.to_numeric(df['Cilindro Pré-op'], errors='coerce')

#df = df.dropna(subset = ['Cilindro Pré-op'])
df = df.reset_index(drop=True)
        
print('Número de olhos com cilindros pré-ops apropriados: ', df.shape[0])

df['Cilindro Pós-op'].value_counts()
for index in range(df.shape[0]):
    val = df['Cilindro Pós-op'][index]
    if isinstance(val, str):
        val = val.strip()
        val = val.replace(',', '.')
        val = val.replace(';', '.')
        if '+' in val:
            df['Cilindro Pós-op'][index] = val[0:val.index('+')]
        if '/' in val:
            df['Cilindro Pós-op'][index] = val[0:val.index('/')]
        if ' ' in val:
            df['Cilindro Pós-op'][index] = val[0:val.index(' ')]

df['Cilindro Pós-op'] = pd.to_numeric(df['Cilindro Pós-op'], errors='coerce')

#df = df.dropna(subset = ['Cilindro Pós-op'])
df = df.reset_index(drop=True)
        
print('Número de olhos com cilindros pós-ops apropriados: ', df.shape[0])
```

```python
#Tratamento dos Eixos
df['Eixo Pré-op'].value_counts()
for index in range(df.shape[0]):
    val = df['Eixo Pré-op'][index]
    if isinstance(val, str):
        val = val.strip()
        val = val.replace(',', '.')
        val = val.replace(';', '.')
        if '+' in val:
            df['Eixo Pré-op'][index] = val[0:val.index('+')]
        if '/' in val:
            df['Eixo Pré-op'][index] = val[0:val.index('/')]
        if ' ' in val:
            df['Eixo Pré-op'][index] = val[0:val.index(' ')]

df['Eixo Pré-op'] = pd.to_numeric(df['Eixo Pré-op'], errors='coerce')

#df = df.dropna(subset = ['Cilindro Pré-op'])
df = df.reset_index(drop=True)
        
print('Número de olhos com eixos pré-ops apropriados: ', df.shape[0])

df['Eixo Pós-op'].value_counts()
for index in range(df.shape[0]):
    val = df['Cilindro Pós-op'][index]
    if isinstance(val, str):
        val = val.strip()
        val = val.replace(',', '.')
        val = val.replace(';', '.')
        if '+' in val:
            df['Eixo Pós-op'][index] = val[0:val.index('+')]
        if '/' in val:
            df['Eixo Pós-op'][index] = val[0:val.index('/')]
        if ' ' in val:
            df['Eixo Pós-op'][index] = val[0:val.index(' ')]

df['Eixo Pós-op'] = pd.to_numeric(df['Eixo Pós-op'], errors='coerce')

#df = df.dropna(subset = ['Cilindro Pós-op'])
df = df.reset_index(drop=True)
        
print('Número de olhos com eixos pós-ops apropriados: ', df.shape[0])
```

```python
#Processamento das Idades
for index in range(df.shape[0]):
    val = df['Idade'][index]
    if isinstance(val, str):
        try:
            df['Idade'][index] = round(((df['Data da cirurgia'][index] - datetime.datetime.strptime(val, '%d/%m/%Y')).days)/365)
        except:
            df = df.drop([index])

df = df.reset_index(drop=True)
df['Idade'] = pd.to_numeric(df['Idade'])

for index in range(df.shape[0]):
    if df['Idade'][index] < 18:
        df = df.drop([index])

df = df.reset_index(drop=True)        
df9 = df

print('Número de pacientes com a idade apropriada e > 18 anos: ', df.shape[0])
```

```python
#Tratamento da Acuidade Pré-Op
for index in range(df.shape[0]):
    val = df['Acuidade Pré-op'][index]
    
    #Tratando as 5 exceções que seriam perdidas sem motivo:
    if val == '2/20':
        val = 2020;
        continue;
    if val == '20/25+2,5':
        val = 2025;
        continue;
    if val == '20/15P':
        val = 2015;
        continue;
    ### FIM DO TRATAMENTO DAS EXCEÇÕES
    
    if isinstance(val, str):
        if '/' in val:
            try:
                val = int(val[0:2] + val[3:])
            except:
                #print('Apagando: ', val)
                #df = df.drop([index])
                continue
        elif val.isdigit():
            val = int(val)
        else:
            #print('Apagando: ', val)
            #df = df.drop([index])
            continue
    df['Acuidade Pós-op'][index] = val

df['Acuidade Pré-op'] = pd.to_numeric(df['Acuidade Pré-op'], errors = 'coerce')

#df = df.dropna(subset=['Acuidade Pré-op'])
df = df.reset_index(drop=True)
df10 = df        
print('Número de olhos com acuidade pré-operatória ajustada: ', df.shape[0])
```

```python
#Limpando novamente pelo filtro de Acuidade pós-op
df['Acuidade Pós-op'].value_counts()
for index in range(df.shape[0]):
    if df['Acuidade Pós-op'][index] > 2040 or df['Acuidade Pós-op'][index] < 2000:
        df = df.drop([index])

df = df.dropna(subset=['Acuidade Pós-op'])
df = df.reset_index(drop=True)

df11 = df
print('Número de olhos com acuidade apropriada: ', df.shape[0])
```

```python
####REMOVER OS OLHOS DUPLICADOS
count = df['Nome'].value_counts()

#dfx = df.copy()

for index in range(df.shape[0]):
    if count[df['Nome'][index]] > 1:
        df = df.drop([index])
        count = df['Nome'].value_counts()

df = df.reset_index(drop = True)
print('Número de olhos finais: ', df.shape[0])
```
